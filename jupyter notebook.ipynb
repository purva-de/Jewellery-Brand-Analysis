```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel(r"C:\Users\purvade\Downloads\Employee_Performance.xlsx")
```

```python
df.head(5)
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df.shape
```

```python
df_cleaned['Join_Date'] = pd.to_datetime(df_cleaned['Join_Date'], errors='coerce')

```

```python
df_cleaned.fillna({
    'Monthly_Sales(INR)': 0,
    'Customer_Feedback_Score': df_cleaned['Customer_Feedback_Score'].mean(),
    'Project_Completion_Rate(%)': df_cleaned['Project_Completion_Rate(%)'].mean(),
    'Attendance(%)': df_cleaned['Attendance(%)'].mean(),
    'Salary(INR)': 0,
    'Bonus(INR)': 0,
    'City': 'Unknown'
}, inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
current_date = datetime.now()
df_cleaned['Experience_Years'] = (current_date - df_cleaned['Join_Date']).dt.days / 365
```

```python
df_cleaned.head(6)
```

```python
df_cleaned['Salary(INR)'] = pd.to_numeric(df_cleaned['Salary(INR)'], errors='coerce')
```

```python
df_cleaned['Total_Earnings(INR)'] = df_cleaned['Salary(INR)'] + df_cleaned['Bonus(INR)']

```

```python
categories = ['High Performer', 'Medium Performer', 'Low Performer']
df_cleaned['Performance_Category'] = pd.cut(df_cleaned['Project_Completion_Rate(%)'], bins=[0, 70, 85, 100], labels=categories, include_lowest=True)
```

```python
df_cleaned.head(6)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_trends = df_cleaned.groupby('Experience_Years')['Monthly_Sales(INR)'].sum().reset_index()
#sales trends
```

```python
dept_feedback = df_cleaned.groupby('Department')['Customer_Feedback_Score'].mean().reset_index()
#dept wise feedback
```

```python
top_performers = df_cleaned[df_cleaned['Performance_Category'] == 'High Performer'][['Name', 'Department', 'Total_Earnings(INR)']]
#top
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=sales_trends, x='Experience_Years', y='Monthly_Sales(INR)', marker="o")
plt.title('Sales Trends Over Employee Experience')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=dept_feedback, x='Department', y='Customer_Feedback_Score')
plt.title('Average Customer Feedback by Department')
plt.xticks(rotation=45)
plt.show()
```
